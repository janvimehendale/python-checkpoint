# 5. Find min, max, and average duration of movies from the 1975-1984.

```python
year_duration_1975_1984 = year_duration[year_duration['release_year'].between(1975, 1984, inclusive = 'both')].sort_values('release_year')
year_duration_1975_1984
```

```python
#year_duration_1975_1984.min()--It is taking min years value that starts with "1" since the data type is object.
```

```python
#year_duration_1975_1984_min = min.(yearyear_duration_1975_1984[["release_year", "duration"]]_duration_1975_1984['duration'])
year_duration_1975_1984_min = (year_duration_1975_1984.loc[
                                year_duration_1975_1984['duration']
                               .str.split() # this is spliting the string
                               .str[0] # it is taking the first element
                               .astype(int) # converting to integer
                               .idxmin(), # location of the minimum value 
                               'duration' # displaying the value of minimum index
                               ]
                              )
year_duration_1975_1984_min
```

```python
year_duration_1975_1984_max = (year_duration_1975_1984.loc[
                                year_duration_1975_1984['duration']
                               .str.split() # this is spliting the string
                               .str[0] # it is taking the first element
                               .astype(int) # converting to integer
                               .idxmax(), # location of the minimum value 
                               'duration' # displaying the value of minimum index
                               ]
                              )
year_duration_1975_1984_max
```

```python
# Finding the sum to calculate the Average mins
year_duration_1975_1984_sum = sum(year_duration_1975_1984['duration']
                                  .str.split()
                                  .str[0]
                                  .astype(int)
                                 )
print(f"{year_duration_1975_1984_sum} min")                   
```

```python
# Finding the count of rows/len to calculate the Average mins
year_duration_1975_1984_len = len(year_duration_1975_1984['duration']
                                  .str.split()
                                  .str[0]
                                  .astype(int)
                                 )
print(f"{year_duration_1975_1984_len} movies")     
```

```python
# Calculating Average
year_duration_1975_1984_avg = year_duration_1975_1984_sum/ year_duration_1975_1984_len
print(f"{round(year_duration_1975_1984_avg,2)} min")
```

# 12a. Which movie had the greatest gap from the year it was released to the year it was added? 
Ans: 	**Steamboat Willie** has **91 years of gap**; released in 1928 and added in 2019

```python
movies_df_year_added = movies_df[["title", "date_added", "release_year"]]
movies_df_year_added
```

```python
movies_df_year_added['date_added'] = pd.to_datetime(movies_df['date_added'])
movies_df_year_added['year_added'] = movies_df_year_added['date_added'].dt.year
movies_df_year_added
```

```python
#movies_df_year_gap = movies_df_year_added[movies_df_year_added["year_added"]-movies_df_year_added["release_year"]].reset_index()
movies_df_year_gap = movies_df_year_added
movies_df_year_gap['gap_years'] = (movies_df_year_gap['year_added']-movies_df_year_gap['release_year'])
movies_df_year_gap.sort_values("gap_years", ascending=False)
```

# 12b. What is the average number of years between release to being added?

```python
movies_df_year_gap_average = sum(movies_df_year_gap['gap_years'])/ len(movies_df_year_gap['gap_years'])
print(f"{round(movies_df_year_gap_average)} years")
```
